```python
import pandas as pd
from IPython.display import Markdown, display
```

```python
df = pd.read_csv("../src/data/Ano-2022.csv", sep=";", low_memory=False)
df.head()
```

```python
columns_data = []

for column in df.columns:
    temp_column = df[column]
    data = temp_column.dropna().value_counts().to_frame().reset_index()[column].to_list()[:3]
    columns_data.append({
        "column": column,
        "data": ", ".join([str(i) for i in data])
    })
```

```python
display(Markdown("## Tabela de Dados"))
display(Markdown(pd.DataFrame(columns_data).to_markdown()))
```

```python
correlation = df["numMes"].corr(df["vlrDocumento"])
print(f"A correlação entre numMes e vlrDocumento é: {correlation}")
```
